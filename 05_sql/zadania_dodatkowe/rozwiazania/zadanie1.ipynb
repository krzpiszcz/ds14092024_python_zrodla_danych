### Zadanie 1

Twoim zadaniem jest wykonanie 18 podzadań, wykorzystując bazę danych company.db z tabelą employees. Poniżej znajdują się zapytania SQL, które należy napisać i wykonać.

Struktura tabeli `employees`



Aby utworzyć bazę `company.db` wykonaj w terminalu, w katalogu w którym znaduje się ten notatnik komendę:
<pre><code>python setup_company_db.py</code></pre>

##### A. Wyświetl imiona i nazwiska wszystkich pracowników.

SELECT first_name, last_name FROM employee;

##### B. Wyświetl imiona i nazwiska pracowników, którzy pracują w dziale 'Engineering'.

SELECT first_name, last_name FROM employee WHERE department = 'Engineering';

##### C. Wyświetl imiona i nazwiska oraz wynagrodzenia pracowników, którzy zarabiają więcej niż 70,000.

SELECT first_name, last_name, salary FROM employee WHERE salary > 70000;

##### D. Wyświetl imiona i nazwiska oraz wiek pracowników, którzy mają mniej niż 30 lat.

SELECT first_name, last_name, age FROM employee WHERE age < 30;

##### E. Wyświetl imiona i nazwiska pracowników zatrudnionych po 1 stycznia 2019 roku.

SELECT first_name, last_name FROM employee WHERE hire_date > '2019-01-01';

##### F. Wyświetl imiona, nazwiska oraz stanowiska pracowników, którzy pracują w dziale 'Marketing' lub 'Sales'.

SELECT first_name, last_name, position FROM employee WHERE department IN ('Marketing', 'Sales');

##### G. Wyświetl imiona, nazwiska oraz wiek pracowników, którzy pracują na stanowisku 'HR Specialist' lub 'HR Manager'.

SELECT first_name, last_name, age FROM employee WHERE position IN ('HR Specialist', 'HR Manager');

##### H. Policz liczbę pracowników w każdym dziale.

SELECT department, COUNT(*) as num_employees FROM employee GROUP BY department;

##### I. Policz średnie wynagrodzenie w każdym dziale.

SELECT department, AVG(salary) as average_salary FROM employee GROUP BY department;

##### J. Policz liczbę pracowników, którzy zarabiają mniej niż 50,000.

SELECT COUNT(*) as num_low_salary FROM employee WHERE salary < 50000;

##### K. Wyświetl imiona, nazwiska i wynagrodzenia trzech najlepiej zarabiających pracowników.

SELECT first_name, last_name, salary FROM employee ORDER BY salary DESC LIMIT 3;

##### L. Wyświetl imiona, nazwiska i daty zatrudnienia pięciu najnowszych pracowników.

SELECT first_name, last_name, hire_date FROM employee ORDER BY hire_date DESC LIMIT 5;

##### M. Wyświetl imiona i nazwiska oraz wynagrodzenia pracowników w dziale 'Finance' posortowane według wynagrodzenia malejąco.

SELECT first_name, last_name, salary FROM employee WHERE department = 'Finance' ORDER BY salary DESC;

##### N. Policz średni wiek pracowników w każdym dziale.

SELECT department, AVG(age) as average_age FROM employee GROUP BY department;

##### O. Wyświetl imiona, nazwiska i wynagrodzenia pracowników, którzy mają więcej niż 40 lat i zarabiają więcej niż 60,000.

SELECT first_name, last_name, salary FROM employee WHERE age > 40 AND salary > 60000;

##### P. Wyświetl imiona, nazwiska i stanowiska pracowników, którzy pracują na stanowisku zawierającym słowo 'Manager'.

SELECT first_name, last_name, position FROM employee WHERE position LIKE '%Manager%';

##### R. Policz łączną sumę wynagrodzeń w każdym dziale.

SELECT department, SUM(salary) as total_salary FROM employee GROUP BY department;

##### S. Wyświetl imiona, nazwiska i wiek pracowników, którzy zostali zatrudnieni w roku 2020 (przykład z gwiazdką).

SELECT first_name, last_name, age FROM employee WHERE strftime('%Y', hire_date) = '2020';
